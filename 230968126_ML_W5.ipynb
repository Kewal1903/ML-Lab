# EXERCISE 1 

Consider the dataset 'weatherAUS' called Rain in Australia. It describes the weather 
characteristics on different dates and locations. This dataset contains about 10 years of daily 
weather observations from many locations across Australia. 

```python
dataset = pd.read_csv("C:\\Users\\mca\\Desktop\\ML_Lab\\Week5\\weatherAUS.csv")
df = pd.DataFrame(dataset)
df
```

1. Preprocess the data, remove the attributes which were are not useful to predict rain. Also, remove rows with at least one missing value for each of them. 

```python
df = df.drop(columns = ['Location','Date'])
```

```python
df = df.dropna()
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df.select_dtypes(include=['object']).columns:
    df[column] = le.fit_transform(df[column])
X = df.drop('Risk_mm', axis=1)
y = df['Risk_mm']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

2. Calculate the Variance Inflation Factor (VIF) value. VIF is a number that determines whether a variable has multicollinearity or not (starts from 1, and it has no upper limit. If the number gets larger, it means the variable has huge multicollinearity on it.)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
numeric_X_train = X_train.select_dtypes(include=[np.number])
numeric_X_train_with_const = add_constant(numeric_X_train)
vif_data = pd.DataFrame()
vif_data["Variable"] = numeric_X_train_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(numeric_X_train_with_const.values, i) for i in range(numeric_X_train_with_const.shape[1])]
high_vif_columns = vif_data[vif_data["VIF"] >= 5]["Variable"].tolist()
high_vif_columns = [col for col in high_vif_columns if col != 'const']
new_features = []
for i in range(len(high_vif_columns)):
    for j in range(i + 1, len(high_vif_columns)):
        feature_1 = high_vif_columns[i]
        feature_2 = high_vif_columns[j]
        new_feature_name = f"{feature_1}_minus_{feature_2}"
        numeric_X_train[new_feature_name] = numeric_X_train[feature_1] - numeric_X_train[feature_2]
        new_features.append(new_feature_name)
numeric_X_test = X_test.select_dtypes(include=[np.number])
numeric_X_test_with_const = add_constant(numeric_X_test)
for i in range(len(high_vif_columns)):
    for j in range(i + 1, len(high_vif_columns)):
        feature_1 = high_vif_columns[i]
        feature_2 = high_vif_columns[j]
        new_feature_name = f"{feature_1}_minus_{feature_2}"
        numeric_X_test[new_feature_name] = numeric_X_test[feature_1] - numeric_X_test[feature_2]
numeric_X_train = numeric_X_train.drop(columns=high_vif_columns)
numeric_X_test = numeric_X_test.drop(columns=high_vif_columns)
```

3. Remove multicollinearities by creating new features. Find the features that have paired values and create the new feature which is the difference value between those pairs. 

```python
numeric_X_train_with_const_new = add_constant(numeric_X_train)
vif_data_new = pd.DataFrame()
vif_data_new["Variable"] = numeric_X_train_with_const_new.columns
vif_data_new["VIF"] = [variance_inflation_factor(numeric_X_train_with_const_new.values, i) for i in range(numeric_X_train_with_const_new.shape[1])]
print("\nUpdated VIF values after creating new features:")
print(vif_data_new)
features_to_remove = vif_data_new[vif_data_new["VIF"] >= 5]["Variable"].tolist()
features_to_remove = [col for col in features_to_remove if col != 'const']
numeric_X_train_cleaned = numeric_X_train.drop(columns=features_to_remove)
numeric_X_test_cleaned = numeric_X_test.drop(columns=features_to_remove)
print("Shape of cleaned dataset:", numeric_X_train_cleaned.shape)
print("Shape of cleaned testing dataset:", numeric_X_test_cleaned.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model_before_vif = LinearRegression()
model_before_vif.fit(X_train_scaled, y_train)
y_pred_before_vif = model_before_vif.predict(X_test_scaled)
mse_before_vif = mean_squared_error(y_test, y_pred_before_vif)
r2_before_vif = r2_score(y_test, y_pred_before_vif)
print("Metrics before VIF computation:")
print(f"Mean Squared Error (MSE): {mse_before_vif:.4f}")
print(f"R-squared: {r2_before_vif:.4f}\n")
```

```python
model_after_vif = LinearRegression()
model_after_vif.fit(numeric_X_train_cleaned, y_train)
y_pred_after_vif = model_after_vif.predict(numeric_X_test_cleaned)
mse_after_vif = mean_squared_error(y_test, y_pred_after_vif)
r2_after_vif = r2_score(y_test, y_pred_after_vif)
print("Metrics after VIF computation:")
print(f"Mean Squared Error (MSE): {mse_after_vif:.4f}")
print(f"R-squared: {r2_after_vif:.4f}\n")
print("Comparison of Models:")
print(f"MSE before VIF: {mse_before_vif:.4f} | MSE after VIF: {mse_after_vif:.4f}")
print(f"R-squared before VIF: {r2_before_vif:.4f} | R-squared after VIF: {r2_after_vif:.4f}")
```
