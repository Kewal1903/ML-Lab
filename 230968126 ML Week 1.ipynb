1.	What are the top 3 sales transactions with the highest revenue?

2.	How many units of each product were sold?

```python
product_counts = df.groupby("Product Name")["Units Sold"].sum()
print("Number of units sold of each product: ")
product_counts
```

3.	What is the total revenue after applying discounts?

```python
revenue_after_discounts = (df["Revenue"]) * (1 - df["Discount Percentage"]/100).sum()
print("Total revenue after applying discounts: ")
revenue_after_discounts
```

4.	Which sales transaction had the highest discount offered, and how much revenue did it generate after applying the discount?

```python
max_discount_transaction = df.loc[df["Discount Percentage"].max()]
max_discount_revenue = max_discount_transaction["Revenue"] * (1 - max_discount_transaction["Discount Percentage"]/100)
print("Revenue after applying maximum discount: ")
max_discount_revenue
```

11.	Which salesperson sold the most units in the North region without offering any discount?

```python
filtered_df = df[(df['Region'] == 'North') & (df['Discount Percentage'] == 0)]

if filtered_df.empty:
    print("No salesperson sold units in the North region without offering any discount.")
else:
    units_sold_per_salesperson = filtered_df.groupby('Salesperson')['Units Sold'].sum()
    max_units_salesperson = units_sold_per_salesperson.idxmax()
    print("The salesperson who sold the most units in the North region without offering any discount is:", max_units_salesperson)
```

12.	What is the average revenue per unit sold in each region for each product?

```python
grouped_df = df.groupby(['Region', 'Product Name']).agg(
    total_revenue = ('Revenue','sum'),
    total_units_sold = ('Units Sold', 'sum')
).reset_index()

grouped_df['avg_revenue_per_unit'] = grouped_df['total_revenue']/grouped_df['total_units_sold']
print(grouped_df[['Region', 'Product Name', 'avg_revenue_per_unit']])
```

13.	Which salesperson has the highest average revenue after discounts, and how does it compare between regions?

```python
df["Revenue After Discount"] = df["Revenue"] * (1 - df["Discount Percentage"]/100)
avg_revenue_per_salesperson_region = df.groupby(["Salesperson", "Region"])["Revenue After Discount"].mean().reset_index()
max_avg_revenue_salesperson = avg_revenue_per_salesperson_region.loc[avg_revenue_per_salesperson_region["Revenue After Discount"].idxmax()]
print("The salesperson with the highest average revenue after discount is: ", max_avg_revenue_salesperson["Salesperson"])
print("Region: ", max_avg_revenue_salesperson["Region"])
print("Average Revenue After Discount :", max_avg_revenue_salesperson["Revenue After Discount"])
```

15.	For each salesperson, rank the transactions by revenue, and find the top 2 transactions for each.

```python
df["Revenue Rank"] = df.groupby("Salesperson")["Revenue"].rank(ascending = False, method = 'first')
top_2_transactions = df[df["Revenue Rank"] <= 2]
print("Top 2 transactions for each salesperson: ")
print(top_2_transactions[["Salesperson", "Date", "Product Name", "Units Sold", "Revenue", "Region", "Discount Percentage"]])
```

16.	How has the total revenue generated by each product changed over time? Show cumulative revenue for each product per day.


```python
df['Date'] = pd.to_datetime(df['Date'])
daily_revenue_per_product = df.groupby(['Product Name', 'Date'])['Revenue'].sum().reset_index()
daily_revenue_per_product['Cumulative Revenue'] = daily_revenue_per_product.groupby('Product Name')['Revenue'].cumsum()
print("Daily revenue per product: ")
print(daily_revenue_per_product[['Product Name', 'Date', 'Cumulative Revenue']])
```

17.	Analyze how discounts affect revenue. For each product, what is the average revenue generated with a discount compared to without a discount? (consider cases where no discount is there for certain products)

```python
with_discount = df[df['Discount Percentage'] > 0]
without_discount = df[df['Discount Percentage'] == 0]
avg_revenue_with_discount = with_discount.groupby('Product Name')['Revenue'].mean()
avg_revenue_without_discount = without_discount.groupby('Product Name')['Revenue'].mean()
avg_revenue_comparison = pd.DataFrame({
    'Avg Revenue with Discount': avg_revenue_with_discount,
    'Avg Revenue without Discount': avg_revenue_without_discount
}).reset_index()
print("Average Revenue Comaparison")
avg_revenue_comparison
```

18.	What is the weighted average discount offered by each salesperson, weighted by the revenue they generated?

```python
df['Weighted Discount'] = df['Revenue'] * df['Discount Percentage']
weighted_avg_discount_per_salesperson = df.groupby('Salesperson').apply(
    lambda x: x['Weighted Discount'].sum() / x['Revenue'].sum()).reset_index(name='Weighted Average Discount')
print("Weighted average discount per salesperson:")
weighted_avg_discount_per_salesperson
```

19.	What percentage of the total revenue does each region contribute? Compare it to the total revenue per region.

```python
total_revenue = df["Revenue"].sum()
regional_revenues = df.groupby("Region")["Revenue"].sum()
percentage_regional_revenues = (regional_revenues/total_revenue)*100
print("Percentage of revenue contributed by each region: ")
percentage_regional_revenues
```
